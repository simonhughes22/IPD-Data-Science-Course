```python
import pylab as pl
import numpy as np
```

```python
from sklearn.cross_validation import train_test_split
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from collections import defaultdict

# create 80%-20% train-test split
def create_datapoints(x_min, x_max, y_min, y_max, data_points=100):
    x_step = (x_max - x_min) / float(data_points)
    y_step = (y_max - y_min) / float(data_points)
    xx, yy = np.meshgrid(np.arange(x_min, x_max, x_step), np.arange(y_min, y_max, y_step))
    X_pred = np.c_[xx.ravel(), yy.ravel()] 
    return X_pred

# utility function to plot the decision surface
def plot_surface(est, x_1, x_2, ax=None, threshold=0.0, contourf=False, plot_boundary=False):
    """Plots the decision surface of ``est`` on features ``x1`` and ``x2``. """
    xx1, xx2 = np.meshgrid(np.linspace(x_1.min(), x_1.max(), 100), 
                           np.linspace(x_2.min(), x_2.max(), 100))
    
    # plot the hyperplane by evaluating the parameters on the grid
    X_pred = np.c_[xx1.ravel(), xx2.ravel()]  # convert 2d grid into seq of points
    if hasattr(est, 'predict_proba'):  # check if ``est`` supports probabilities
        # take probability of positive class
        pred = est.predict_proba(X_pred)[:, 1]
    else:
        pred = est.predict(X_pred)
    Z = pred.reshape((100, 100))  # reshape seq to grid
    if ax is None:
        ax = plt.gca()
    # plot line via contour plot
    if contourf:
        ax.contourf(xx1, xx2, Z, levels=np.linspace(0, 1.0, 10), cmap=plt.cm.GnBu, alpha=0.6)
    
    if plot_boundary:
        ax.contour(xx1, xx2, Z, levels=[threshold], linewidths=2.0, colors='green')
    ax.set_xlim((x_1.min(), x_1.max()))
    ax.set_ylim((x_2.min(), x_2.max()))

def add_scatter(ax, x1, x2, clss, dct_clr, alpha=1.0):
    cls2data = defaultdict(list)
    for r_ix in range(clss.shape[0]):
        cls = clss[r_ix]
        xx1 = x1[r_ix]
        xx2 = x2[r_ix]
        cls2data[cls].append((xx1,xx2,cls))

    for key, lst in cls2data.items():
        color = dct_clr[key]
        xx1,xx2,cls = zip(*lst)
        if key < 0:
            ax.scatter(xx1, xx2, s=100, c=color, marker='o')
        else:
            #ax.scatter(xx1, xx2, s=10,   c=color)
            pass
    
def plot_dataset(X_train, X_test, y_train, y_test, est = None, fn = None, lbl = "", plot_boundary=False):

    fig, (ax, ax2) = plt.subplots(1, 2, figsize=(30, 10))
    
    if fn is not None:
        xmin = X_train[:,0].min()
        xmax = X_train[:,0].max()
        xx1s = np.linspace(xmin,xmax,1000)
        #TODO plot function
        points = fn(xx1s)
        xys = zip(xx1s, points)
        xys = [(x,y) for x,y in xys if y >= xmin and y <= xmax]
        xs,ys = zip(*xys)
        ax.plot(xs,ys,  color='black', linewidth=2)
        ax2.plot(xs,ys, color='black', linewidth=2)
        #ax3.set_xlim( X_train[:, 0].min(), X_train[:, 0].max())
        #ax3.set_ylim(X_train[:, 1].min(), X_train[:, 1].max())
        #ax3.set_title("Target Function", fontsize=25)
        pass
    
    y_train = y_train.copy()
    y_test  = y_test.copy()
    
    # plot decision surface, add in error points
    if est is not None:
        est.fit(X_train, y_train)
        
        t_predict = est.predict(X_train)
        err = (y_train != t_predict).mean()
        y_train[y_train != t_predict] = -1.0

        plot_surface(est, X_train[:, 0], X_train[:, 1], ax=ax, threshold=0.5, contourf=True, plot_boundary=plot_boundary)
        #ax.text(0.88, 0.02, '%.2f' % err, transform=ax.transAxes)
        ax.set_xlabel('Training Error: %.2f' % err, fontsize=25)
 
        v_predict = est.predict(X_test)
        y_test[y_test != v_predict] = -1.0

        plot_surface(est, X_test[:, 0], X_test[:, 1], ax=ax2, threshold=0.5, contourf=True, plot_boundary=plot_boundary)
        err2 = (y_test != v_predict).mean()
        #ax2.text(0.88, 0.02, '%.2f' % err2, transform=ax2.transAxes)
        ax2.set_xlabel('Test Error:    %.2f' % err2, fontsize=25)

    # Plot the training points
    dct2clr = { 0: "green", 1 : "purple", -1: "#FF3333" }
    # Plot the errors only
    add_scatter(ax,  X_train[:, 0], X_train[:, 1], y_train, dct2clr)
    # and testing points
    add_scatter(ax2, X_test[:, 0],  X_test[:, 1],  y_test,  dct2clr, alpha=0.6)
    # plot limits
    ax.set_xlim( X_train[:, 0].min(), X_train[:, 0].max())
    ax2.set_ylim(X_train[:, 1].min(), X_train[:, 1].max())

    ax.set_ylabel('$x_1$')
    ax2.set_ylabel('$x_1$')
    
    if est is not None:
        classifier = str(type(est))[1:-1].replace("class ", "")[1:-1].split(".")[-1]
        ax.set_title(lbl  + " " + classifier, fontsize=25)
        ax2.set_title(lbl + " " + classifier, fontsize=25)
    else:
        ax.set_xlabel('$x_0$')
        ax2.set_xlabel('$x_0$')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

x1_min, x1_max, x2_min, x2_max = -10, 10, -10, 10
def plot_fn(fn, est, data_points=100, lbl="", plot_fn=True, X=None, Y=None, plot_boundary=False):
    def fn_as_class(x):
        x1 = x[:,0]
        x2 = x[:, 1]
        rvals = fn(x1)
        clsvals = rvals <= x2
        return clsvals.astype(np.float)

    if X is None:
        X = create_datapoints(x1_min, x1_max, x2_min, x2_max, data_points=data_points)
    if Y is None:
        Y = fn_as_class(X)
    # create train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5)
    if plot_fn:
        plot_dataset(X_train, X_test, y_train, y_test, est=est, fn=fn, lbl=lbl,   plot_boundary=plot_boundary)
    else:
        plot_dataset(X_train, X_test, y_train, y_test, est=est, fn=None, lbl=lbl, plot_boundary=plot_boundary)
```

Linear Function + Linear Classifier
===================================

```python
def linear_fn(x, grad=1.0, intercept=2.0):
    return grad * x + intercept

data_points = 30
plot_fn(fn=linear_fn, est=LogisticRegression(),     data_points=data_points, plot_fn=True)
#plot_fn(fn=linear_fn, est=DecisionTreeClassifier(), data_points=data_points)
```

Linear Separability
===================

```python
def non_linearly_separable(x):
    ys = []
    threshold = 5.25
    for i in range(x.shape[0]):
        x1 = x[i, 0]
        x2 = x[i, 1]
        y = 1 if (x1 >=threshold and x2 <=threshold) or (x1 <= threshold and x2 >= threshold) else 0
        ys.append(y)
    return np.asarray(ys)

data_points = 50
X = create_datapoints(0, x1_max, 0, x2_max, data_points=data_points)
Y = non_linearly_separable(X)

plot_fn(fn=None, est=LogisticRegression(), X=X, Y=Y, plot_fn=False, plot_boundary=True)
plot_fn(fn=None, est=DecisionTreeClassifier(), X=X, Y=Y, plot_fn=False)
plot_fn(fn=None, est=KNeighborsClassifier(n_neighbors=1),  X=X, Y=Y, plot_fn=False, lbl="1")

# A linear classifier (such as logistic regression), cannot learn a function that's not linearly separable 
# (even if regions are split by straight lines)
```

Non-Linear Function - Linear vs Non-Linear Classifier
=====================================================

```python
def nonlinear_fn(x):
    return x**2 - 0.25*x**3 - 2.0

data_points = 50
X = create_datapoints(x1_min, x1_max, x2_min, x2_max, data_points=data_points)
plot_fn(fn=nonlinear_fn, est=LogisticRegression(),     data_points=data_points, X=X)
plot_fn(fn=nonlinear_fn, est=DecisionTreeClassifier(), data_points=data_points, X=X)
plot_fn(fn=nonlinear_fn, est=KNeighborsClassifier(n_neighbors=1), data_points=data_points, X=X)
```

Noisy Data - Overfitting and Regularization
===========================================

```python
data_points = 20
X = create_datapoints(x1_min, x1_max, x2_min, x2_max, data_points=data_points)

def noisy_non_linear_fn(x):
    # fix random number generation so repeatable
    np.random.seed(0)
    return x**2 + 0.25*x**3 + np.random.normal(0,0.5,x.shape[0])

plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=1,  weights="distance"), data_points=data_points, lbl="1", X=X)
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=1,  weights="distance"), data_points=data_points, lbl="1",  plot_fn=False, X=X)
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=3,  weights="distance"), data_points=data_points, lbl="3" , plot_fn=False, X=X) 
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=3,  weights="distance"), data_points=data_points, lbl="5" , plot_fn=False, X=X) 
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=20, weights="distance"), data_points=data_points, lbl="15", plot_fn=False, X=X) 

#SH : First plot with function, rest without to show decision boundary
#SH : Most datasets contain noise (measurement error, human error, etc)
#SH : When Powerful classifiers overfit, they aren't able to generalize well and distinguish signal from noise)
#SH : Regularization methods learn smoother and typically simpler functions. For Knn, higher values of k (#neighbors) 
#SH :   mean smoother, simpler functions are modelled.
```

More Data Points (200 vs 20) Reduces Overfitting (Knn)
=====================================================

```python
data_points = 200
X = create_datapoints(x1_min, x1_max, x2_min, x2_max, data_points=data_points)

plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=1,  weights="distance"), data_points=data_points, lbl="1",  plot_fn=False, X=X)
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=3,  weights="distance"), data_points=data_points, lbl="3",  plot_fn=False, X=X)
plot_fn(fn=noisy_non_linear_fn, est=KNeighborsClassifier(n_neighbors=15, weights="distance"), data_points=data_points, lbl="15", plot_fn=False, X=X)

# SH : With 10 X data, one neighbor is sufficient for same error, additional regularization (i.e. neighbors) don't help
# SH : Note how the decision surface gets smoother as k increases
```
